```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def load_image():
    blank_image = np.zeros((600,800))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_image,text='Jenittan',org=(50,300),fontFace=font,fontScale = 5,color=(255,255,255),thickness=15,lineType=cv2.LINE_AA)
    return blank_image
```

```python

img=load_image()
```

```python
plt.title("Output") 
plt.axis('off')
plt.imshow(img)
```

```python
kernel=np.ones((5,5),dtype=np.uint8)
```

```python
erosion = cv2.erode(img,kernel,iterations = 2)
```

```python
plt.title("Eroded Image") 
plt.axis('off')
plt.imshow(erosion)
```

```python
dilution=cv2.dilate(img,kernel,iterations=2)
plt.title("Dilated Image") 
plt.axis('off')
plt.imshow(dilution)
```
